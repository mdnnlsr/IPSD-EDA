```python
# Visualize unique values

#menghitung data unik
unique_counts = data.nunique()
#membuat bar plot
unique_counts.plot(kind='bar', figsize=(8,6), color='pink')
#memasukkan judul untuk bar plot
plt.title("persebaran unique values")
#menampilkan bar plot
plt.show()
```

```python
# Filter data for 'Field_of_Study' as 'Science' and 'Gender' as 'Female'
filtered_data = data[(data['Field_of_Study'] == 'Science') & (data['Gender'] == 'Female')]

# mengatur ukuran  gambar
plt.figure(figsize=(7, 6))
# menghitung jumlah observasi pada sumbu x
sns.countplot(x='Grade_Level', data=filtered_data, color='salmon')
# menambahkan judul
plt.title("Count of Female Students in Science by Grade Level")
plt.xlabel("Grade Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter data for 'Field_of_Study' as 'Engineering' and 'Gender' as 'Male'
filtered_data2 = data[(data['Field_of_Study'] == 'Engineering') & (data['Gender'] == 'Male')]

plt.figure(figsize=(7, 6))
sns.countplot(x='Grade_Level', data=filtered_data2, color='salmon')
plt.title("Count of Female Students in Engineering by Grade Level")
plt.xlabel("Grade Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Box plot for numeric columns 

# mengatur ukuran
plt.figure(figsize=(15,8))
# kolom yang dipilih
sns.boxplot(data=data[['Age', 'Perceived_Effectiveness_of_VR', 'Impact_on_Creativity', 'Engagement_Level', 'Hours_of_VR_Usage_Per_Week']])
# menambahkan judul
plt.title("Box Plot of Numeric Columns")
# menampilkan
plt.show()
```

```python
# Select the specified columns for correlation
columns_for_correlation = ['Age', 'Perceived_Effectiveness_of_VR', 'Impact_on_Creativity', 
                           'Engagement_Level', 'Hours_of_VR_Usage_Per_Week']

# Compute the correlation matrix
correlation_matrix = data[columns_for_correlation].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix of Selected Columns")
plt.show()
```
